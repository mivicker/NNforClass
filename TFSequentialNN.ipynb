```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
from helpel import *
```

```python
model = tf.keras.Sequential([keras.layers.Dense(50, input_shape=[1], activation = "relu"),
                            keras.layers.Dense(50, activation = "relu"),
                            keras.layers.Dense(50, activation = "relu"),
                            keras.layers.Dense(1)])
model.compile(optimizer = "rmsprop", loss = "mse")
```

```python
X, y = generate_data()

plt.plot(X[:, 0], y, '*')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
model.fit(X[:,0], y, epochs=5000, verbose=False)
```

```python
preds = model.predict(X[:,0])
```

```python
plt.plot(X[:, 0], y)
plt.plot(X[:, 0], preds)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
